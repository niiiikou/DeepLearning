```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def get_parm(r1,r2,i,m):
    a = max(0,i-r1)
    b = min (m,i+r1+1)
    c = min (i+r2+1,m)
    d = max(0,i-r2)
    return a,b,c,d

def activate(input):
    if input < 0:
        return 0
    elif 0 <= input < 2:
        return input
    else:
        return 2
```

```python
# step1 - initialize parameteres and weights
signal = [0.27, 0.35, 0.44, 0.58, 0.66, 0.77, 0.4, 0.32, 0.20, 0.15, 0.08]
r1 = 0
r2 = np.inf
c1 = 3
c2 = -0.4
m = len(signal)
x = signal.copy()
# step 2
t_max = 3
x_max = 2
for t in range(0, t_max):#step 3 step 7 & 8
    x_old = x.copy() 
    plt.plot(x_old,label ="t "+ str(t))
    plt.legend()
    print("-------------------------------------------------------")
    print ('t = {}'.format(t))
    print("-------------------------------------------------------")
    for i in range(0, m):
        a,b,c,d = get_parm(r1, r2, i, m)
        x[i] = c1 * sum((x_old[a:b])) +c2 * sum((x_old[b :c])) + c2 * sum((x_old[d: a]))
        x[i] = activate(round((x[i]), 3)) 
        print ('x{} = {}'.format(i, x[i]))

plt.show()
```
