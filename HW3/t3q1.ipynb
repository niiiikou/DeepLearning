```python
def plot_out(input_1,input_2,input_3,output_1,output_2,output_3,w,t):
  if t == (1,1):
    st_1 = np.outer(input_1, output_1)
    st_2 = np.outer(input_2, output_2)
    st_3 = np.outer(input_3, output_3)
    w = st_1 + st_2 + st_3

  test_1 = np.where(np.inner(input_1.reshape((1,63)), w.T) > 0, 1, -1)
  test_2 = np.where(np.inner(input_2.reshape((1,63)), w.T) > 0, 1, -1)
  test_3= np.where(np.inner(input_3.reshape((1,63)), w.T) > 0, 1, -1)
  plt.subplot(3, 3, 1)
  plt.imshow(input_1.reshape((9,7)),cmap="gray")
  plt.title('input')
  plt.subplot(3, 3, 2)
  plt.imshow(output_1.reshape(t),cmap="gray")
  plt.title('output')
  plt.subplot(3, 3, 3)
  plt.imshow(test_1.reshape(t),cmap="gray")
  plt.title('predicted_output')
  plt.subplot(3, 3, 4)
  plt.imshow(input_2.reshape((9,7)),cmap="gray")
  plt.subplot(3, 3, 5)
  plt.imshow(output_2.reshape(t),cmap="gray")
  plt.subplot(3, 3, 6)
  plt.imshow(test_2.reshape(t),cmap="gray")
  plt.subplot(3, 3, 7)
  plt.imshow(input_3.reshape((9,7)),cmap="gray")
  plt.subplot(3, 3, 8)
  plt.imshow(output_3.reshape(t),cmap="gray")
  plt.subplot(3, 3, 9)
  plt.imshow(test_3.reshape(t),cmap="gray")
  return test_1,test_2,test_3
```

```python
def replaceRandom(arr, perc):
    temp = arr   
    shape = temp.shape[0]      
    noise = round((perc * shape)/100 )
    temp = temp.flatten()    # Flatten to 1D
    inds = np.random.randint(0,shape,noise)   # Get random indices
    # Fill with something
    for i in inds:
      if arr[i] == 1:
        temp[i] = -1
      else:
        temp[i] = 1
    return temp
```

```python
def noisy_input(input_1,input_2,input_3,output_1,output_2,output_3,noise,random_func,w,t):
  input_1_noise = random_func(input_1, noise)
  input_2_noise = random_func(input_2, noise)
  input_3_noise = random_func(input_3, noise)
  
  test_1,test_2,test_3 = plot_out(input_1_noise,input_2_noise,input_3_noise,output_1,output_2,output_3,w,t)
  return test_1,test_2,test_3
```

```python
def perc(input_1,input_2,input_3,output_1,output_2,output_3,noise,func,w,itr,t):
  c1 = 0
  c2 = 0
  c3 = 0
  for i in range(itr):
    test_1,test_2,test_3 = noisy_input(input_1,input_2,input_3,output_1,output_2,output_3,noise,func,w,t)
    if (test_1 == output_1).all():
      c1+=1
    if (test_2 == output_2).all():
      c2+=1
    if (test_3 == output_3).all():
      c3+=1
  print('percentage_output1 : ',c1*100/itr)
  print('percentage_output2 : ',c2*100/itr)
  print('percentage_output3 : ',c3*100/itr)
```

# output_size : (5,3)

```python
t =(5,3)
```

```python
perc(input_1,input_2,input_3,output_1,output_2,output_3,20,replaceRandom,w,500,t)
```

```python
perc(input_1,input_2,input_3,output_1,output_2,output_3,60,replaceRandom,w,500,t)
```
