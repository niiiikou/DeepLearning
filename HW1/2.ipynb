```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

# الف

```python
#class 1 data points
mean_class1 = (0, 0)
cov_class1 = [[0.09, 0], [0, 0.09]]
class1 = np.random.multivariate_normal(mean_class1, cov_class1, 100)
class1_with_labels =np.c_[ class1, np.ones(class1.shape[0]) ] 
#class 2 data points
mean_class2 = (2, 2)
cov_class2 = [[0.09, 0], [0, 0.09]]
class2 = np.random.multivariate_normal(mean_class2, cov_class2, 100)
class2_with_labels =np.c_[ class2,-1* np.ones(class2.shape[0]) ] 
#combine two classes 
data = np.r_[ class1_with_labels, class2_with_labels ]
#shuffle 
np.random.shuffle(data) 
#plot
plt.plot(class1_with_labels[:,0], class1_with_labels[:,1], 'ro', label='class1')
plt.plot(class2_with_labels[:,0], class2_with_labels[:,1], 'bo', label='class2')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title('data points')
plt.show()
```

# ب

```python
# x and y
x = data[:,[0,1]]
# target
target = data[:,2]

alpha = 0.001
epoch = 100
#weight
w = [0, 0] 
#bias
b = 0  
cost_iteration = []
for j in range(epoch):
    errors = 0
    cost = 0
    for i in range(x.shape[0]):
        t = target[i]
        net = (x[i,0]*w[0]) + (x[i,1]*w[1]) + b 
        # Update the weights and the bias term  
        w[0] = w[0] + (alpha*(t-net)*x[i,0])    
        w[1] = w[1] + (alpha*(t-net)*x[i,1])
        b= b + (alpha*(t-net))
        error_of_each_data = (t -net)**2
        errors += error_of_each_data
    #np.array(errors)
    cost = errors * 0.5
    #print('cost :', cost)
    cost_iteration.append(cost)

# error plot
plt.plot(range(1, len(cost_iteration) + 1), cost_iteration)
plt.xlabel('Epochs')
plt.ylabel('Sum-squared-error')
plt.title('Adeline ')
plt.show()
```

```python
# sepratetor line and datapoints plot 
plt.plot(class1_with_labels[:,0], class1_with_labels[:,1], 'ro', label='class1')
plt.plot(class2_with_labels[:,0], class2_with_labels[:,1], 'bo', label='class2')
x = np.arange(-1, 4)
y =((w[0]*x)+b)/(-w[1])
plt.plot(x, y, "g--")
plt.legend()
plt.title("datapoints with Seperator line")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
